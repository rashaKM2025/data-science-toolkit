```python
# 1. Create a numpy array of size 2 x 2
import numpy as np
array = np.array([[1, 2],[3, 4]])
print(array)
```

```python
# 4. Create a random numpy array of size 6x3 and transpose it
array = np.array([[5,3,2],[2,8,0],[9,1,5],[8,0,4],[4,2,6],[0,9,3]])
print(array)

print()

transpose = array.T
print(transpose)
```

```python
array = np.random.randint(0,9, size=(6, 3))
print(array)
print()
transpose = array.T
print(transpose)
```

```python
# 5. Create two random numpy array of size n x p and p x m and then compute the dot product
n, p = 2, 3
array = np.array([[2,5,6],[9,8,7],])
print(array)
print()
p, m = 3, 2
array2 = np.array([[2,3],[6,7],[9,0]])
print(array2)
print()
dot = np.dot(array, array2)
print(dot)
```
